```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/iot_device_status.csv")
```

```python
df
```

```python
x  = df.iloc[:,:-1]
y = df.iloc[:,-1:]
```

```python
#split into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Instantiate the KNN model
knn = KNeighborsClassifier(n_neighbors=5) # You can adjust n_neighbors

# Train the model, converting y_train to a 1D array
knn.fit(x_train, y_train.values.ravel())
```

```python
# Make predictions on the test set
y_pred = knn.predict(x_test)
y_pred

```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
# Compute the confusion matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
#accuracy
accuracy = accuracy_score(y_test, y_pred)
accuracy_percentage = accuracy * 100
print("accuracy percentage is :",accuracy_percentage,"%")
```

```python
#save model
import pickle
filename = "knn_model.pkl"
pickle.dump(knn,open(filename,"wb"))
```

```python
loaded_model = pickle.load(open(filename,"rb"))
predicted_status = loaded_model.predict([[2000, 0.1, 100, 45]])
print(predicted_status)
```

```python
#Classification Report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
